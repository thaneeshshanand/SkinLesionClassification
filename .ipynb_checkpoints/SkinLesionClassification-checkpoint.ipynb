```python
img = cv2.imread("ISIC-images/ISIC_0024310.jpg")
```

```python
img_rbg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rbg)
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray)
plt.axis("off")
plt.show()
```

```python
 # Create a kernel for morphological operations
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (9, 9))
```

```python
blackhat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.imshow(blackhat)
plt.axis("off")
plt.show()
```

```python
_, mask = cv2.threshold(blackhat, 10, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(mask)
plt.axis("off")
plt.show()
```

```python
result = cv2.inpaint(img, mask, 6, cv2.INPAINT_TELEA)
```

```python
plt.imshow(result)
plt.axis("off")
plt.show()
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
image2 = cv2.GaussianBlur(result, (5, 5), 0)
```

```python
plt.imshow(image2)
plt.axis("off")
plt.show()
```

```python
image3 = image2.astype(np.float32) / 255.0
```

```python
plt.imshow(image3)
plt.axis("off")
plt.show()
```
